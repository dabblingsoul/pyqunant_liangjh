```python
# largest constituents of SPY
symbols = openbb.etf.holdings('SPY')
spy_symbols = list(symbols.index.drop(['N/A']))
spy_valuation = openbb.stocks.ca.screener(similar=spy_symbols, data_type='valuation')
spy_valuation = spy_valuation.sort_values(by=['Price'], ascending=False).convert_dtypes()
spy_valuation.head(10)
```

```python
spy_daily = openbb.stocks.load(
    symbol='SPY', start_date='1993-11-01', monthly=True
)

openbb.stocks.candle(
    data=spy_daily, asset_type='SPY - Monthly Chart from 1993-11', symbol=''
)
```
